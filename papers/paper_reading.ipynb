### Monocular Depth Estimation using Diffusion Models

In my previous understanding, diffusion models are typically highly coupled with image generation tasks, where the convention of a diffusion model is to gradually generate a structurally sound and semantically clear image through a step-by-step denoising process from pure Gaussian noise. Using a diffusion model was somewhat beyond my current intuitive understanding of this type of models, urging me to take a closer look on how this paper achieves SOTA performance of this field.

project link: https://depth-gen.github.io/

#### KEY TAKEAWAYS

In addition to providing a new SOTA solution to depth estimation task by , I draw thhe following inspiration from this paper:

To begin with, this paper proposed a promising way to mitigate data quality and quantity issues and potentially generate additional useful training signals by leveraging the strong generative capability of diffusion model. As they noted that "much of the existing data is noisy and incomplete," causing problems like "a problematic distribution shift between training and testing." This might be a new direction in the field of depth estimation. 

Last but not least, this paper noted a difference between training and actual inference of the model. Therefore, their introduced a method named Step-Unrolled Denoising Diffusion. This hightlights the importannce of narrowing the gap between training and inference as much as possible as it might leed to more desirable performance

### Genie 3: A new frontier for world models

Later this week, I encountered some practical difficulties when when attempting to reconstruct 3D hand states using hand pose annotation and monocular depth estimation, espeically monocular depth estimation via UniDepth. These challenges motivated to come up with an alternative way to get the vital data to reconstruct the world state: to simply generate a virtual world where all data are explicitly accessible so that we can bypass such challenge. This led to my explorarion on world models, the models runs on physical and spatial laws.

paper link: https://deepmind.google/blog/genie-3-a-new-frontier-for-world-models/

<img src="img/worldmodel.png">

#### KEY TAKEAWAYS:


As highlighted by Google, world models represent a highly promising research direction, where we can synthesize a plausible interactive world that serves as the basis of downstream embodied intelligence tasks. From my persepctive, such a workflow has the potential to greatly simplify data collection and environment construction, enabling more efficient and scalable training of embodied agents. 

### QUESTIONS  REMAINING

As inference-time efficiency remains a critical concern for real-world embodied systems, an critical challenge is how current mainstream methods balance accuracy and latency in practice.


Another question concerns the role of simulation in embodied intelligence: to what extent can learned world models serve as substitutes for traditional physics-based simulators or explicit real-world reconstruction pipelines?
